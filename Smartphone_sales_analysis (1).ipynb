**Smartphone sales analysis in India**

```python
highest_rated = data.sort_values(by=["Rating"], ascending=False)
highest_rated = highest_rated.head(10)
print(highest_rated[['Brands', 'Discount']])
```

**Let's a look at the number of ratings of the highest rated phone**

```python

plt.bar(highest_rated['Brands'], highest_rated['Rating'])
plt.xlabel("Brands")
plt.ylabel("Rating")
plt.title("Ratings of highest_rated phones")
plt.show()
```

**Discount of highest_rated phones**

```python
plt.bar(highest_rated['Brands'], highest_rated['Discount'])
plt.xlabel("Brands")
plt.ylabel("Discount")
plt.title("Discount of highest_rated phones")
plt.show()
```

**Relationship between Selling price and Rating**

```python
data_frame = pd.DataFrame(data)
x = data_frame['Rating']
y = data_frame['Selling Price']
plt.scatter(x, y, c="r")
plt.xlabel("Rating")
plt.ylabel("Selling Price")
plt.title("Relationship between Selling Price and rating")
plt.show()
```
